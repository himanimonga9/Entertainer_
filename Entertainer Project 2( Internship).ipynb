```python
combined_df['Year of Breakthrough/#1 Hit/Award Nomination'] = pd.to_numeric(combined_df['Year of Breakthrough/#1 Hit/Award Nomination'], errors='coerce')
combined_df['Birth Year'] = pd.to_numeric(combined_df['Birth Year'], errors='coerce')
combined_df['Breakthrough Age'] = combined_df['Year of Breakthrough/#1 Hit/Award Nomination'] - combined_df['Birth Year']
print(combined_df)
```

```python
combined_df
```

```python
combined_df['Year of First Oscar/Grammy/Emmy'] = pd.to_numeric(combined_df['Year of First Oscar/Grammy/Emmy'], errors='coerce')
combined_df['Year of Breakthrough/#1 Hit/Award Nomination'] = pd.to_numeric(combined_df['Year of Breakthrough/#1 Hit/Award Nomination'], errors='coerce')
combined_df['Difference of getting highly prestigious award after breakthrough'] = combined_df['Year of First Oscar/Grammy/Emmy'] - combined_df['Year of Breakthrough/#1 Hit/Award Nomination']
print(combined_df)
```

```python
combined_df.drop_duplicates(inplace = True)
combined_df.shape
```

```python
combined_df
```

```python
combined_df['Year of Last Major Work (arguable)'] = pd.to_numeric(combined_df['Year of Last Major Work (arguable)'], errors='coerce')
combined_df['Year of Breakthrough/#1 Hit/Award Nomination'] = pd.to_numeric(combined_df['Year of Breakthrough/#1 Hit/Award Nomination'], errors='coerce')
combined_df['Worked(years) after breakthrough'] = combined_df['Year of Last Major Work (arguable)'] - combined_df['Year of Breakthrough/#1 Hit/Award Nomination']
print(combined_df)
```

```python
replaced_columns = ['Year of First Oscar/Grammy/Emmy','Year of Last Major Work (arguable)', 'Difference of getting highly prestigious award after breakthrough']

column_averages = combined_df[replaced_columns].apply(pd.to_numeric, errors='coerce').mean()
for column in replaced_columns:
    combined_df[column] = combined_df[column].fillna(column_averages[column])
```

```python
combined_df.to_csv('updated_dataset.xlsx', index=False)
```
